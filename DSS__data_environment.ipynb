```python
# Update the graph to ensure the arrow from SOCRATES to Whaikaha is one-directional

# Create a directed graph
G = nx.DiGraph()

# Add nodes with their roles
G.add_node("Whaikaha", role="Government Agency")
G.add_node("SOCRATES", role="Data Management System")
G.add_node("NASCs", role="Needs Assessment & Coordination")
G.add_node("Manawanui", role="IF Services Provider")
G.add_node("Service Providers", role="Service Delivery")
G.add_node("IDI", role="Integrated Data Infrastructure")
G.add_node("Other Gov Agencies", role="Government Agencies")
G.add_node("Stats NZ", role="Stats NZ")
G.add_node("Other Researchers", role="Other Researchers")

# Add edges to represent data flow
G.add_edges_from([
    ("SOCRATES", "Whaikaha"),  # Ensure one-directional arrow
    ("SOCRATES", "NASCs"),
    ("NASCs", "Service Providers"),
    ("Service Providers", "Manawanui"),
    ("NASCs", "Manawanui"),
    ("Service Providers", "SOCRATES"),
    ("SOCRATES", "IDI"),
    ("SOCRATES", "Other Gov Agencies"),
    ("Other Gov Agencies", "IDI"),
    ("Stats NZ", "IDI"),
    ("IDI", "Whaikaha"),
    ("IDI", "Manawanui"),
    ("IDI", "Other Researchers")
])

# Define node positions to avoid overlap
pos = {
    "Whaikaha": (0, 2),
    "SOCRATES": (2, 2),
    "NASCs": (2, 1),
    "Service Providers": (2, 0),
    "Manawanui": (3, 1),
    "IDI": (4, 2),
    "Other Gov Agencies": (1, 3),
    "Stats NZ": (3, 3),
    "Other Researchers": (5, 1)
}

# Define node sizes based on their roles
node_sizes = [3000 if G.nodes[node]['role'] == "Data Management System" else 2000 for node in G]

# Define node labels with their roles
node_labels = {
    "Whaikaha": "Whaikaha\n(Government Agency)",
    "SOCRATES": "SOCRATES\n(Data Management System)",
    "NASCs": "NASCs\n(Assessment & Coordination)",
    "Manawanui": "Manawanui\n(IF Services Provider)",
    "Service Providers": "Service Providers\n(Service Delivery)",
    "IDI": "IDI\n(Integrated Data Infrastructure)",
    "Other Gov Agencies": "Other Gov Agencies\n(Government Agencies)",
    "Stats NZ": "Stats NZ",
    "Other Researchers": "Other Researchers"
}

# Draw the updated graph with roles displayed and improved aesthetics
plt.figure(figsize=(14, 10))
nx.draw_networkx(G, pos, labels=node_labels, with_labels=True, node_size=node_sizes, node_color="skyblue", font_size=10, font_color="black", font_weight="bold", arrows=True, arrowstyle='-|>', arrowsize=12)

# Ensure text doesn't overlap
plt.tight_layout()

# Adjust text positions to avoid overlap
for _, spine in plt.gca().spines.items():
    spine.set_visible(False)
plt.xticks([])
plt.yticks([])

plt.title("Current Data Flow in Disability Support Services Ecosystem", fontsize=15)
plt.show()
```

```python
import matplotlib.pyplot as plt
import networkx as nx

# Create a directed graph
G = nx.DiGraph()

# Add nodes with their roles
G.add_node("Whaikaha", role="Government Agency")
G.add_node("Central Data Hub", role="Central Data Repository")
G.add_node("NASCs", role="Needs Assessment & Coordination")
G.add_node("Manawanui", role="IF Services Provider")
G.add_node("Service Providers", role="Service Delivery")
G.add_node("IDI", role="Integrated Data Infrastructure")
G.add_node("Other Gov Agencies", role="Government Agencies")
G.add_node("Stats NZ", role="Stats NZ")
G.add_node("Other Researchers", role="Other Researchers")
G.add_node("Analytics Platform", role="Analytics and Reporting")

# Add edges to represent data flow
G.add_edges_from([
    ("Whaikaha", "Central Data Hub"),
    ("NASCs","Central Data Hub"),
    ("NASCs", "Service Providers"),
    ("Service Providers", "Central Data Hub"),
    ("Manawanui", "Central Data Hub"),
    ("Other Gov Agencies", "Central Data Hub"),
    ("Other Gov Agencies", "IDI"),
    ("Stats NZ", "IDI"),
    ("IDI", "Manawanui"),
    ("IDI", "Other Researchers"),
    ("Central Data Hub", "IDI"),
    ("Central Data Hub", "Analytics Platform"),
    ("Analytics Platform", "Whaikaha"),
    ("Analytics Platform", "NASCs"),
    ("Analytics Platform", "Service Providers"),
    ("Analytics Platform", "Manawanui")
])

# Define node positions to avoid overlap
pos = {
    "Whaikaha": (0, 2),
    "Central Data Hub": (2, 2),
    "NASCs": (2, 1),
    "Service Providers": (1, 0),
    "Manawanui": (3, 1),
    "IDI": (4, 2),
    "Other Gov Agencies": (1, 3),
    "Stats NZ": (3, 3),
    "Other Researchers": (5, 1),
    "Analytics Platform": (3, 2.5)
}

# Define node sizes based on their roles
node_sizes = [3000 if G.nodes[node]['role'] == "Central Data Repository" else 2000 for node in G]

# Define node labels with their roles
node_labels = {
    "Whaikaha": "Whaikaha\n(Government Agency)",
    "Central Data Hub": "Central Data Hub\n(SOCRATRES + Other Data Repository)",
    "NASCs": "NASCs\n(Assessment & Coordination)",
    "Manawanui": "Manawanui\n(IF Services Provider)",
    "Service Providers": "Service Providers\n(Service Delivery)",
    "IDI": "IDI\n(Integrated Data Infrastructure)",
    "Other Gov Agencies": "Other Gov Agencies\n(Government Agencies)",
    "Stats NZ": "Stats NZ",
    "Other Researchers": "Other Researchers",
    "Analytics Platform": "Analytics Platform\n(Reporting & Insights)"
}

# Draw the updated graph with roles displayed and improved aesthetics
plt.figure(figsize=(16, 12))
nx.draw_networkx(G, pos, labels=node_labels, with_labels=True, node_size=node_sizes, node_color="skyblue", font_size=10, font_color="black", font_weight="bold", arrows=True, arrowstyle='-|>', arrowsize=12)

# Ensure text doesn't overlap
plt.tight_layout()

# Adjust text positions to avoid overlap
for _, spine in plt.gca().spines.items():
    spine.set_visible(False)
plt.xticks([])
plt.yticks([])

plt.title("Ideal Data Flow in Disability Support Services Ecosystem", fontsize=15)
plt.show()
```
